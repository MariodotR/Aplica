1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 

# graficar
plt.plot(
    df['month_number'], # eje x
    df['total_profit'],# eje y
    color="red", # color
    linewidth=1 # tamano de la curva
)
plt.title("Gráfico de lineas") # agregar titulo
plt.xlabel('Nº mes') # nombre eje x
plt.ylabel('Beneficio total') # nombre eje y
plt.grid(color='w') # agregar grillado
plt.show() # mostrar grafico
```

```python
fig = plt.figure(figsize=(10, 5)) # ventana

'''
ax = plt.axes()
# Cambiando el color de fondo
ax.set_facecolor("w")
'''
plt.scatter(
    df['month_number'], # eje x
    df['total_profit'],# eje y
    c="r", # color
    alpha=0.9, # intensidad color
    label="Beneficios Totales" # etiquetdo fijando posicion
)
plt.xlabel('Nº mes') # nombre eje x
plt.title("Gráfico de dispersión")
plt.legend(loc='upper left')
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
lista_colores= ["red","black","blue","green","yellow"]
# tamano del grafico
fig = plt.figure(figsize=(10, 5)) 
for i in range(5):
    
    plt.plot(
    df['month_number'], # eje x
    df.iloc[:,i], # eje y
    label=df.columns[i], # etiquetado
    color=lista_colores[i-1], # color
    linewidth=1 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.xlabel('Nº mes') # nombre eje x
plt.ylabel('ventas mensuales') # nombre eje y
plt.grid() # agregar grillado
plt.show() # mostrar grafico
```
